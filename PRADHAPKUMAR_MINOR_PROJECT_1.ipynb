```python
# ============================================
# GROUPDNA - MINOR PROJECT
# PART 1 : CHAT PARSER
# ============================================

# Read the WhatsApp chat file

file = open("hostel_bois.txt", "r", encoding="utf-8")
lines = file.readlines()
file.close()


# List to store all messages
messages = []


# Counters
system_messages = 0
media_messages = 0
deleted_messages = 0


# Read every line one by one
for line in lines:

    # Remove extra spaces and new line
    line = line.strip()

    # Skip empty line
    if line == "":
        continue

    # Check whether the line has timestamp
    if " - " not in line:
        continue

    # Split timestamp and remaining text
    parts = line.split(" - ", 1)

    timestamp = parts[0]
    remaining = parts[1]

    # System message has no sender
    if ": " not in remaining:
        system_messages += 1
        continue

    # Split sender and message
    sender, text = remaining.split(": ", 1)

    # Check media message
    is_media = False
    if text == "<Media omitted>":
        media_messages += 1
        is_media = True

    # Check deleted message
    is_deleted = False
    if text == "This message was deleted":
        deleted_messages += 1
        is_deleted = True

    # Store everything inside dictionary
    chat = {
        "timestamp": timestamp,
        "sender": sender,
        "message": text,
        "media": is_media,
        "deleted": is_deleted
    }

    messages.append(chat)


# ============================================
# BASIC OUTPUT
# ============================================

print("=" * 55)
print("CHAT PARSER COMPLETED")
print("=" * 55)

print("Total Parsed Messages :", len(messages))
print("System Messages       :", system_messages)
print("Media Messages        :", media_messages)
print("Deleted Messages      :", deleted_messages)

print()


# Display first 5 messages

print("FIRST 5 PARSED MESSAGES")
print("-" * 55)

for i in range(5):
    print(messages[i])
```

```python
# ============================================
# GROUPDNA - MINOR PROJECT
# PART 2 : GROUP OVERVIEW
# ============================================

from datetime import datetime

# Dictionary to store message count of each person
person_count = {}

# Set to store unique participants
participants = set()

# List to store all dates
dates = []

# Loop through all parsed messages
for chat in messages:

    sender = chat["sender"]

    # Count messages
    if sender in person_count:
        person_count[sender] += 1
    else:
        person_count[sender] = 1

    # Add participant
    participants.add(sender)

    # Convert timestamp into datetime format
    dt = datetime.strptime(chat["timestamp"], "%d/%m/%y, %H:%M")
    dates.append(dt)

# Find first and last date
first_date = min(dates)
last_date = max(dates)

# Total number of days
total_days = (last_date - first_date).days + 1

# Total messages
total_messages = len(messages)

# Total participants
total_participants = len(participants)

# Sort people by message count (Highest to Lowest)
sorted_people = sorted(person_count.items(), key=lambda x: x[1], reverse=True)

# ============================================
# PRINT REPORT
# ============================================

print("=" * 60)
print("               GROUP OVERVIEW")
print("=" * 60)

print("Group Name        : Hostel Bois 4ever")
print("Period            :", first_date.strftime("%d %B %Y"), "to", last_date.strftime("%d %B %Y"))
print("Total Days        :", total_days)
print("Total Messages    :", total_messages)
print("Participants      :", total_participants)

print("\nParticipants List")
print("-" * 60)

for name in sorted(participants):
    print(name)

print("\nMessages Per Person")
print("-" * 60)

for name, count in sorted_people:

    percentage = (count / total_messages) * 100

    print(f"{name:10} : {count:4} messages ({percentage:.1f}%)")
```

```python
# ============================================
# GROUPDNA - MINOR PROJECT
# PART 3 : MOST ACTIVE DAY & HOUR
# ============================================

from datetime import datetime

# Dictionary to count messages for each day
day_count = {}

# Dictionary to count messages for each hour
hour_count = {}

# Loop through all messages
for chat in messages:

    # Convert timestamp into datetime
    dt = datetime.strptime(chat["timestamp"], "%d/%m/%y, %H:%M")

    # Get date only
    day = dt.strftime("%d %B %Y")

    # Get hour only
    hour = dt.hour

    # Count messages for each day
    if day in day_count:
        day_count[day] += 1
    else:
        day_count[day] = 1

    # Count messages for each hour
    if hour in hour_count:
        hour_count[hour] += 1
    else:
        hour_count[hour] = 1


# ============================================
# Find busiest day
# ============================================

busiest_day = ""
highest_day_messages = 0

for day in day_count:

    if day_count[day] > highest_day_messages:
        highest_day_messages = day_count[day]
        busiest_day = day


# ============================================
# Find busiest hour
# ============================================

busiest_hour = 0
highest_hour_messages = 0

for hour in hour_count:

    if hour_count[hour] > highest_hour_messages:
        highest_hour_messages = hour_count[hour]
        busiest_hour = hour


# Average messages per day during that hour
average = highest_hour_messages / total_days


# ============================================
# PRINT RESULT
# ============================================

print("=" * 60)
print("         MOST ACTIVE DAY & HOUR")
print("=" * 60)

print(f"Busiest Day  : {busiest_day}")
print(f"Messages     : {highest_day_messages}")

print()

print(f"Busiest Hour : {busiest_hour:02d}:00 - {(busiest_hour+1)%24:02d}:00")
print(f"Messages     : {highest_hour_messages}")
print(f"Average/Day  : {average:.2f}")

print()

print("=" * 60)
print("TOP 5 BUSIEST DAYS")
print("=" * 60)

# Sort days
top_days = sorted(day_count.items(), key=lambda x: x[1], reverse=True)

for day, count in top_days[:5]:
    print(f"{day:20} {count} messages")

print()

print("=" * 60)
print("TOP 5 BUSIEST HOURS")
print("=" * 60)

# Sort hours
top_hours = sorted(hour_count.items(), key=lambda x: x[1], reverse=True)

for hour, count in top_hours[:5]:

    print(f"{hour:02d}:00 - {(hour+1)%24:02d}:00   {count} messages")
```

```python
# ============================================
# GROUPDNA - MINOR PROJECT
# PART 4 : ACTIVITY HEATMAP (NUMPY)
# ============================================

import numpy as np
from datetime import datetime

# Get all participant names
people = sorted(list(participants))

# Create dictionary for row number
person_index = {}

for i in range(len(people)):
    person_index[people[i]] = i

# Create a 6 x 24 matrix filled with zeros
heatmap = np.zeros((len(people), 24), dtype=int)

# Fill the matrix
for chat in messages:

    sender = chat["sender"]

    dt = datetime.strptime(chat["timestamp"], "%d/%m/%y, %H:%M")

    hour = dt.hour

    row = person_index[sender]

    heatmap[row][hour] += 1


# ============================================
# PRINT NUMPY MATRIX
# ============================================

print("=" * 90)
print("ACTIVITY MATRIX")
print("=" * 90)

print("Rows :", people)
print("Columns : Hours (0 - 23)\n")

print(heatmap)


# ============================================
# TEXT HEATMAP
# ============================================

print("\n")
print("=" * 90)
print("ACTIVITY HEATMAP")
print("=" * 90)

# Print hour headings

print("Name      ", end="")

for h in range(24):
    print(f"{h:02}", end=" ")

print()

print("-" * 90)

# Print each person's activity

for i in range(len(people)):

    print(f"{people[i]:10}", end="")

    maximum = max(heatmap[i])

    if maximum == 0:
        maximum = 1

    for value in heatmap[i]:

        percentage = (value / maximum) * 100

        if percentage == 0:
            symbol = ". "

        elif percentage <= 25:
            symbol = "‚ĖĎ "

        elif percentage <= 50:
            symbol = "‚Ėí "

        else:
            symbol = "‚Ėą "

        print(symbol, end="")

    print()


# ============================================
# TOTAL MESSAGES EACH HOUR
# ============================================

print("\n")
print("=" * 90)
print("TOTAL GROUP ACTIVITY BY HOUR")
print("=" * 90)

hour_total = np.sum(heatmap, axis=0)

for h in range(24):
    print(f"{h:02}:00 - {(h+1)%24:02}:00 : {hour_total[h]} messages")


# ============================================
# TOTAL MESSAGES BY EACH PERSON
# ============================================

print("\n")
print("=" * 90)
print("TOTAL MESSAGES PER PERSON")
print("=" * 90)

person_total = np.sum(heatmap, axis=1)

for i in range(len(people)):
    print(f"{people[i]:10} : {person_total[i]} messages")
```

```python
# ============================================
# GROUPDNA - MINOR PROJECT
# PART 5 : TOP 10 WORDS
# ============================================

# Dictionary to store word frequency
word_count = {}

# Common stop words
stop_words = [
    "i","me","my","we","our","you","your","he","she","it","they",
    "is","am","are","was","were","be","been","being",
    "the","a","an","and","or","to","of","in","on","for",
    "at","with","this","that","these","those","as","by","from",
    "have","has","had","do","does","did","will","would",
    "can","could","should","not","so","if","but","it's","its",
    "im","i'm","yaar","hai"
]

# Characters to remove
punctuation = ".,!?;:()[]{}<>\"'/-_@#$%^&*+="

# Loop through every message
for chat in messages:

    # Skip media messages
    if chat["media"] == True:
        continue

    # Skip deleted messages
    if chat["deleted"] == True:
        continue

    message = chat["message"].lower()

    # Remove punctuation
    for p in punctuation:
        message = message.replace(p, "")

    # Split sentence into words
    words = message.split()

    # Count words
    for word in words:

        if word in stop_words:
            continue

        if len(word) <= 1:
            continue

        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1


# ============================================
# SORT WORDS
# ============================================

top_words = sorted(word_count.items(),
                   key=lambda x: x[1],
                   reverse=True)


# ============================================
# PRINT TOP 10 WORDS
# ============================================

print("="*60)
print("        TOP 10 MOST USED WORDS")
print("="*60)

highest = top_words[0][1]

for word, count in top_words[:10]:

    # Create text bar
    bar_length = int((count / highest) * 30)

    bar = "‚Ėą" * bar_length

    print(f"{word:<15} {bar} {count}")


# ============================================
# TOTAL UNIQUE WORDS
# ============================================

print()
print("="*60)

print("Total Unique Words :", len(word_count))

print("="*60)
```

```python
# ============================================
# GROUPDNA - MINOR PROJECT
# PART 6 : RESPONSE SPEED & SILENT STREAK
# ============================================

from datetime import datetime, timedelta

# Dictionary to store response times
response_time = {}

# --------------------------------------------
# Calculate Response Time
# --------------------------------------------

for i in range(1, len(messages)):

    previous = messages[i - 1]
    current = messages[i]

    # Ignore if same person sends consecutive messages
    if previous["sender"] == current["sender"]:
        continue

    # Convert timestamps
    previous_time = datetime.strptime(previous["timestamp"], "%d/%m/%y, %H:%M")
    current_time = datetime.strptime(current["timestamp"], "%d/%m/%y, %H:%M")

    # Find time difference
    difference = current_time - previous_time

    minutes = difference.total_seconds() / 60

    person = current["sender"]

    if person not in response_time:
        response_time[person] = []

    response_time[person].append(minutes)


# --------------------------------------------
# Print Average Response Time
# --------------------------------------------

print("=" * 70)
print("AVERAGE RESPONSE TIME")
print("=" * 70)

for person in sorted(response_time):

    average = sum(response_time[person]) / len(response_time[person])

    print(f"{person:10} : {average:.2f} minutes")


# ============================================
# LONGEST SILENT STREAK
# ============================================

print()
print("=" * 70)
print("LONGEST SILENT STREAK")
print("=" * 70)

# Get first and last date
start_date = min(dates).date()
end_date = max(dates).date()

for person in sorted(participants):

    # Store active dates
    active_days = set()

    for chat in messages:

        if chat["sender"] == person:

            dt = datetime.strptime(chat["timestamp"], "%d/%m/%y, %H:%M")

            active_days.add(dt.date())

    current_day = start_date

    streak = 0
    longest = 0

    while current_day <= end_date:

        if current_day not in active_days:

            streak += 1

            if streak > longest:
                longest = streak

        else:
            streak = 0

        current_day = current_day + timedelta(days=1)

    print(f"{person:10} : {longest} days")


# ============================================
# FASTEST REPLIER
# ============================================

print()
print("=" * 70)

fastest_person = ""
fastest_time = 999999

for person in response_time:

    average = sum(response_time[person]) / len(response_time[person])

    if average < fastest_time:
        fastest_time = average
        fastest_person = person

print("Fastest Replier")
print("----------------")
print(f"{fastest_person} ({fastest_time:.2f} minutes)")
```

```python
# ============================================
# GROUPDNA - MINOR PROJECT
# PART 7 : PERSONALITY ARCHETYPES
# ============================================

from datetime import datetime

# Dictionary to store statistics of each person
personality = {}

# Initialize dictionary
for person in participants:

    personality[person] = {
        "messages": 0,
        "night": 0,
        "question": 0,
        "long_messages": 0,
        "media": 0
    }

# --------------------------------------------
# Collect statistics
# --------------------------------------------

for chat in messages:

    person = chat["sender"]

    personality[person]["messages"] += 1

    # Night messages (10 PM to 5 AM)
    dt = datetime.strptime(chat["timestamp"], "%d/%m/%y, %H:%M")

    if dt.hour >= 22 or dt.hour < 5:
        personality[person]["night"] += 1

    # Question messages
    if "?" in chat["message"]:
        personality[person]["question"] += 1

    # Long messages
    words = chat["message"].split()

    if len(words) >= 15:
        personality[person]["long_messages"] += 1

    # Media messages
    if chat["media"] == True:
        personality[person]["media"] += 1


# ============================================
# Find Maximum Values
# ============================================

max_messages = max(personality, key=lambda x: personality[x]["messages"])
max_night = max(personality, key=lambda x: personality[x]["night"])
max_question = max(personality, key=lambda x: personality[x]["question"])
max_long = max(personality, key=lambda x: personality[x]["long_messages"])
max_media = max(personality, key=lambda x: personality[x]["media"])


# ============================================
# PRINT PERSONALITY REPORT
# ============================================

print("=" * 70)
print("           PERSONALITY ARCHETYPES")
print("=" * 70)

print(f"ūüďĪ Spammer       : {max_messages}")
print(f"ūüĆô Night Owl     : {max_night}")
print(f"‚Ěď Curious Mind  : {max_question}")
print(f"ūüďĖ Story Teller  : {max_long}")
print(f"ūüĖľÔłŹ Media King    : {max_media}")

print()

print("=" * 70)
print("DETAILED STATISTICS")
print("=" * 70)

for person in sorted(participants):

    print(f"\nPerson : {person}")

    print(f"Messages      : {personality[person]['messages']}")
    print(f"Night Msgs    : {personality[person]['night']}")
    print(f"Questions     : {personality[person]['question']}")
    print(f"Long Messages : {personality[person]['long_messages']}")
    print(f"Media Shared  : {personality[person]['media']}")
```
